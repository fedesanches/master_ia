¿Qué tan bien puede una combinación lineal de las funcione ex, sin(x) and Γ(x) aproximar la función f(x) = x−1 en el intervalo [1, 2]? ¿y en el intervalo [0, 1]? Responda ambas preguntas discretizando los respectivos intervalos y realizando un ajuste por cuadrados mínimos.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Discretización del intervalo [1,2)
# Genera 100 puntos del 1 al 2, con el 2 no incluido por default
xx = np.linspace(1, 2, num = 100)
xx
```

```python
yy = xx**-1
yy
```

```python
# Grafiquemos la funcion exponencial
plt.plot(xx, np.exp(xx), 'b-', label='$e^x$')

# Grafiquemos la funcion seno
plt.plot(xx, np.sin(xx), 'g--', label='$\sin(x)$')

# Grafiquemos la funcion gamma
from scipy.special import gamma
plt.plot(xx, gamma(xx), 'r-.', label='$\Gamma(x)$')

# le pedimos que escriba las leyendas y ejes
plt.legend()
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
```

La aproximacion linea seria de la forma:
    f(x) = a*e^x + b*sin(x) + c*gamma(x)

```python
A = np.array([[np.exp(x), np.sin(x), gamma(x)] for x in xx])
A
```

```python
a, b, c = np.linalg.solve(A.T @ A, A.T @ yy)
print(a, b, c)

plt.plot(xx, xx**-1, color="blue")
plt.plot(xx, a*np.exp(xx) + b*np.sin(xx) + c*gamma(xx), color="red")
```
