Dados los tiempos t = (0, 1, 3, 4) y las mediciones b = (1, 8, 8, 20).
    a) Encontrar la recta y = c + dt que mejor se ajusta a los datos. Para esto escriba las ecuaciones normales y resuélvalas. Calcular el error.
    b) Repetir con una línea recta horizontal y = c.
    c) Repetir con una línea recta que pasa por el origen y = dt.
    d) Repetir con una parábola y = a + bt + ct2

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.array([0, 1, 3, 4])
b = np.array([1, 8, 8, 20])

plt.plot(t, b, "bo")
```

a) y = dt+ c
Para esto hay que armar el sistema de matrizes
    y1 = dt1 + c
    y2 = dt2 + c
    y3 = dt3 + c
    y4 = dt4 + c
Que seria
    y1   =  t1  1      d
    y2      t2  1      c
    y3      t3  1
    y4      t4  1

Ecuacion normal = AT * Ax = AT * y

```python
A = np.array([[x, 1] for x in t])
A
```

```python
# Para resolver las ecuaciones normales utilizamos la funcion de numpy para resolver sistemas lineales
d, c = np.linalg.solve(A.T @ A, A.T @ b)

plt.plot(t, b, "bo")
plt.plot(t, d*t + c, "g-")
```

```python
# Calculamos los errores --> 1/2||y - Ax||^2
errores = []
for ii in range(4):
  err = (b[ii] - (d*t[ii] + c))**2
  errores.append(err)

print(errores)

error_total = 0.5*np.sum(errores)
error_total
```

b) Repetir con una línea recta horizontal y = c


```python
A = np.array([[1] for x in t])
A
```

```python
c = np.linalg.solve(A.T @ A, A.T @ b)
print(c)

plt.plot(t, b, "bo")
plt.plot(np.full_like(np.linspace(min(t) - 1, max(t) + 1, 5), c))
```

```python
# Calculamos los errores --> 1/2||y - Ax||^2
errores = []
for ii in range(4):
  err = (b[ii] - (c))**2
  errores.append(err)

print(errores)

error_total = 0.5*np.sum(errores)
error_total
```

Repetir con una línea recta que pasa por el origen y = dt

```python
A = np.array([[x] for x in t])
A
```

```python
d = np.linalg.solve(A.T @ A, A.T @ b)
print(d)

plt.plot(t, b, "bo")
plt.plot(t, d*t, "g-")
```

```python
# Calculamos los errores --> 1/2||y - Ax||^2
errores = []
for ii in range(4):
  err = (b[ii] - (d*t[ii]))**2
  errores.append(err)

print(errores)

error_total = 0.5*np.sum(errores)
error_total
```

Repetir con una parábola y = a + dt + ct2

```python
A = np.array([[x**2, x, 1] for x in t])
A
```

```python
c, d, a = np.linalg.solve(A.T @ A, A.T @ b)
print(c, d, a)

plt.plot(t, b, "bo")
plt.plot(t, c*(t**2) + d*t + a, "g-")
```

```python
# Calculamos los errores --> 1/2||y - Ax||^2
errores = []
for ii in range(4):
  err = (b[ii] - (c*(t[ii]**2) + d*t[ii] + a))**2
  errores.append(err)

print(errores)

error_total = 0.5*np.sum(errores)
error_total
```
