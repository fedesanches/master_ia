```python
def contiene_duplicados(a):
    a.sort()
    print(a)
    for i in range(1, len(a)):
        if a[i] == a[i - 1]:
            return True
    return False

a = [0,3,5,8,1,1]

print(contiene_duplicados(a))
```

```python
a = ['hola', 'c', 'si', 'a', 'no', 'b', 'chau', 'z']

def cadena_ordenada(a, k):
    buckets = [[] for v in range(k + 1)]
    for x in a:
        tamao_palabra = len(x)
        buckets[tamao_palabra].append(x)
    
    b = []
    for x in range(k + 1):
        for i in range(len(buckets[x])):
            b.append(buckets[x][i])
    return b

print(cadena_ordenada(a, 4))
```

```python
a = [0,3,5,8,1,3]

def ordenar_enteros_con_dif(a):
    min_val = min(a)
    max_val = max(a)
    d = max_val - min_val
    
    b = [0 for i in range(d + 1)]
    for x in a:
        b[x] = b[x] + 1

    j = 0
    for x in range(d + 1):
        for i in range(b[x]):
            a[j] = x
            j += 1
    return a

print(ordenar_enteros_con_dif(a))
```

```python
a = [0,3,5,1,5,3]

def ordernar_en_on(a):
    b = [0 for i in range(len(a))]
    for x in a:
        b[x] += 1

    j = 0
    for x in range(len(b)):
        for i in range(b[x]):
            a[j] = x
            j += 1
    return a

print(ordernar_en_on(a))

""" 
En este caso especfico, el algoritmo es efectivamente O(n) porque:
    El rango de valores est acotado por n
    Cada elemento se procesa exactamente una vez en cada paso
    Los bucles anidados finales en total procesan exactamente n elementos
"""
```

```python
def radix_sort_inplace(A):
    n = len(A)

    # -------- primera pasada: ordenar por r = x % n --------
    count = [0] * n
    for x in A:
        count[x % n] += 1
    for i in range(1, n):
        count[i] += count[i-1]

    output = [0] * n
    for x in reversed(A):
        r = x % n
        count[r] -= 1
        output[count[r]] = x
    for i in range(n):
        A[i] = output[i]   # copiar de vuelta a A

    print(count)
    print(A)

    # -------- segunda pasada: ordenar por q = x // n --------
    count = [0] * n
    for x in A:
        count[x // n] += 1
    for i in range(1, n):
        count[i] += count[i-1]

    output = [0] * n
    for x in reversed(A):
        q = x // n
        count[q] -= 1
        output[count[q]] = x
    for i in range(n):
        A[i] = output[i]   # copiar de vuelta a A
    
    print(count)
    print(A)

    return A

A = [7, 0, 15, 3, 12, 4, 8]  # n = 7  rango [0, 49]

print(radix_sort_inplace(A)) 
```
