```python
a = [-2, 9, 10, 11, 13, 14, 15, 17, 19, 25, 50, 51, 49, 29, 12, 8, 7, 5]

def pico(a):
    if len(a) == 0: return 0
    
    i = 0
    j = len(a) - 1
    while i + 1 < j:
        m = (i + j) // 2
        if a[m] < a[m + 1]:
            i = m + 1
        else:
            j = m

    return a[i]

print(pico(a))
```

```python
def merge(a, b):
    # Esto es de la forma O(n*k^2)
    merge = []
    i, j = 0, 0
    
    while i < len(a) and j < len(b):
        if a[i] < b[j]:
            merge.append(a[i])
            i += 1
        else:
            merge.append(b[j])
            j += 1
            
    while i < len(a):
        merge.append(a[i])
        i += 1

    while j < len(b):
        merge.append(b[j])
        j += 1
    
    return merge
```
